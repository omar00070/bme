```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('DataFromImages.xlsx', header=None)
df.columns = ['values']
```

```python
# n, bins, patches = df.plot.hist('values', bins=8)
def plot_hist(column, title):
    fig = plt.figure(figsize=(16,10))
    _, bins, _ = plt.hist(column)
    plt.xticks(bins)
    plt.title(title)
    plt.show()
plot_hist(df.values, 'Histogram of Values')
```

```python
# freqs
from collections import defaultdict
def get_freq_table(filename):
    with open(filename) as file:
        freq_table = defaultdict(int)
        text = next(file)
        for word in text.split(' '):
            word = _preprocess_word(word)
            freq_table[word] += 1
    return freq_table
```

```python
def _preprocess_word(word):
    #remove any apostrophe
    word = word.split("'")[0]
    return ''.join(letter for letter in word if letter.isalnum()).lower()
# print(_preprocess_word("!@#something")
assert(_preprocess_word("!@#sOMething!") == "something")
```

```python
df_freq = pd.DataFrame(get_freq_table("Word_to_Frequencies.txt"), index=["frequency"])
```

```python
df_freq.transpose().plot(kind="bar", figsize=(20, 10), title="Word Frequency Bar Chart")
```
